```python
def normalFunS(x, mu, sd):
    a = 1/(sd*np.sqrt(2*np.pi))
    b = np.exp(-(x-mu)**2/(2*sd**2))
    return(a*b)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

x = np.array(inpeople["wifi사용자수"])
n, bins, ig = plt.hist(x, bins = 43, normed=True)
mu = np.mean(x)
sd = np.std(x)
plt.plot(bins, normalFunS(bins, mu, sd), linewidth = 2)
plt.show()
```
